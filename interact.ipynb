```python
import os
import json

import pandas as pd
import numpy as np


# read tools-for-thought.csv
df = pd.read_csv('tools-for-thought.csv')
```

```python
df.shape
```

```python
# extract clean_text and id
docs = df[['clean_text', 'id']]

base_json = {
    "id": "",
    "text": ""
}

documents = []

# loop through docs and create json
for index, row in docs.iterrows():
    doc = base_json.copy()
    doc['id'] = row['id']
    doc['text'] = row['clean_text']
    documents.append(doc)

```

```python
import requests

# send a post request to the endpoint http://0.0.0.0:8000/upsert


# make documents a json
json_docs = json.dumps({"documents": documents})

# bearer token is 1234
headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Bearer 1234'
}

# send the request
response = requests.post("http://0.0.0.0:8000/upsert", headers=headers, data=json_docs)

response
```

```python
# query

query = {
  "queries": [
    {
      "query": "",
      "top_k": 3
    }
  ]
}

# make query a json
query = json.dumps(query)

response = requests.post("http://0.0.0.0:8000/query", headers=headers, data=query)

response.json()
```
